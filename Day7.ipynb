```python
a, b, c = list(map(int, input().split()))
if a == b:
  print("YES")
else:
  if c > 0 and b < a:
    print("NO")
  elif c < 0 and b > a:
    print("NO")
  elif c == 0:
    print("NO")
  else:
    if (b - a) % c == 0:
      print("YES")
    else:
      print("NO")
```

```python
n, m = list(map(int, input().split()))
ori = []
cmprs = []
diff = []
for i in range(n):
  a, b = list(map(int, input().split()))
  ori.append(a)
  cmprs.append(b)
  diff.append(a-b)
s = sum(ori)
diff.sort(reverse=True)
if sum(cmprs) > m:
  print(-1)
else:
  for j in range(n+1):
    if s <= m:
      print(j)
      break
    if s > m:
      s -= diff[j]
```

```python
def f(x):
  y = x + 1
  while y % 10 == 0:
    y = y // 10
  return y

n = int(input())
d = len(str(n))
out = set()
out.add(n)
for i in range(10*d):
  a = f(n)
  out.add(a)
  n = a
print(len(out))
```

```python
n, m = list(map(int, input().split()))
can = [0]*n
for i in range(m):
  c = list(map(int, input().split()))
  index = c.index(max(c))
  can[index] += 1
print(can.index(max(can)) + 1)
```

```python
t = int(input())
for i in range(t):
  n, T = list(map(int, input().split()))
  a = list(map(int, input().split()))
  b = [-1]*n
  checker = 0
  for j in range(n):
    if a[j] < T/2:
      b[j] = '0'
    elif a[j] > T/2:
      b[j] = '1'
    else:
      if checker % 2 == 0:
        b[j] = '0'
      else:
        b[j] = '1'
      checker += 1
  print(' '.join(b))
```
