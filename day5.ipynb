```python
x, y, z, t1, t2, t3 = list(map(int, input().split()))
# elevator
te = abs(z-x)*t2 + 3*t3 + abs(y-x)*t2
# staircase
ts = abs(y-x)*t1
if ts < te:
  print("NO")
else:
  print("YES")
```

```python
n = int(input())
words = list(input().split())
volumes = []
for w in words:
  volumes.append(len([l for l in w if l.isupper()]))
print(max(volumes))
```

```python
p, n= list(map(int, input().split()))
d=dict()
conflict = -1
for x in range(0, 301):
    d[x]='empty'
for i in range(n):
  entry = int(input())
  rm = entry % p
  if d[rm] == 'empty':
    d[rm] = entry
  else:
    conflict = 1
    print(i+1)
    break
if conflict == -1:
  print(-1)
```
