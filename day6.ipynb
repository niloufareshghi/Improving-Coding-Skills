```python
t = input()
x = list(map(int, input().split()))
for xi in x:
  if xi < 15:
    print("NO")
    continue
  p = 0
  for i in range(15,21):
    if (xi - i) % 14 == 0:
      p = 1
      break
  if p == 1:
    print("YES")
  else:
    print("NO")
```
