```python
n = int(input())
ws = []
hs = []
pixels = []
for i in range(n):
  wi, hi = list(map(int, input().split()))
  ws.append(wi)
  hs.append(hi)
s = sum(ws)
m1 = sorted(hs, reverse=True)[0]
m2 = sorted(hs, reverse=True)[1]
for i in range(n):
  W = s - ws[i]
  if hs[i] == m1:
    H = m2
  else:
    H = m1
  pixels.append(str(W * H))
print(' '.join(pixels))
```

```python
## O(n^2)
n = int(input())
seq = list(map(int, input().split()))
smpl = list(range(n))
for i in range(n):
  if seq  == smpl:
    print("Yes")
    break
  for j in range(n):
    if j % 2 == 0:
      seq[j] += 1
    else:
      seq[j] -= 1
    if seq[j] > n - 1:
      seq[j] = 0
    if seq[j] < 0:
      seq[j] = n - 1
if seq != smpl:
  print("No")
```

```python
## O(n)
n = int(input())
seq = list(map(int, input().split()))
smpl = list(range(n))
cnt = 0
for i in range(n):
  if seq[0] == smpl[0]:
    cnt = i
    break
  seq[0] += 1
  if seq[0] > n - 1:
    seq[0] = 0
for i in range(1, n):
  if i % 2 == 0:
    seq[i] += cnt
  else:
    seq[i] -= cnt
  if seq[i] > n - 1:
    seq[i] -= n - 1
  if seq[i] < 0 :
    seq[i] += n
print(seq)
```

```python
n = int(input())
t = int(input())
a1, b1 = list(map(int, input().split()))
c1 = 7 - a1
d1 = 7 - b1
ls = []
ls.append(sorted([a1, b1, c1, d1]))
for i in range(n-1):
  ai, bi = list(map(int, input().split()))
  ci = 7 - ai
  di = 7 - bi
  l = sorted([ai, bi, ci, di])
  ls.append(l)
s = set(tuple(i) for i in ls)
if len(s) == 1:
  print("YES")
else:
  print("NO")
```
