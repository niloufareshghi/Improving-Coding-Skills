```python
w = int(input())
if w == 1 or w == 2:
  print("NO")
else:
  if w % 2 == 0:
    print("YES")
  else:
    print("NO")
```

```python
n = int(input())
for i in range(n):
  w = input()
  if len(w) < 11:
    print(w)
  else:
    print(w[0] + str(len(w) - 2) + w[-1])
```

```python
n = int(input())
q = 0
for i in range(n):
  ls = [int(item) for item in input().split()]
  if sum(ls) > 1:
    q = q + 1
print(q)
```

```python
n, k = [int(x) for x in input().split()]
ls = [int(item) for item in input().split()]
cnt  = 0
kscore = ls[k - 1]
for i in range(n):
  if ls[i] >= kscore and ls[i] > 0:
    cnt += 1
print(cnt)
```
