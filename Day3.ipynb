```python
t = int(input())
for i in range(t):
  n = int(input())
  ls = [int(j) for j in input().split()]
  res = 0
  for k in range(n - 1):
    res = max(res, ls[k]*ls[k+1])
  print(res)
```

```python
t = int(input())
for i in range(t):
  steps = 0
  n = int(input())
  a = [int(j) for j in input().split()]
  if 0 in a:
    indices = [index for index, value in enumerate(a) if value == 0]
    for id in indices:
      a[id] += 1
      steps += 1
  while sum(a) == 0:
    a[0] += 1
    steps += 1
  print(steps)
```

```python
t = int(input())
for i in range(t):
  a, b = list(map(int, input().strip().split())) 
  x = a & b
  ans = (a ^ x) + (b ^ x)
  print(ans)
```

```python
n = int(input())
e = input()
s = ['0'] * 10
for event in e:
  if event == 'L':
    for i in range(n):
      if s[i] == '0':
        s[i] = '1'
        break
  else:
    if event == 'R':
      for i in range(n):
        if s[9 - i] == '0':
          s[9- i] = '1'
          break
    else:
      s[int(event)] = '0'
print("".join(s))
```

```python
t = int(input())
for i in range(t):
  sentence = input()
  # KOREAN 'mnida' suffix
  if sentence[-5:] == 'mnida':
    print("KOREAN")
  # FILIPINO 'po' suffix
  if sentence[-2:] == 'po':
    print("FILIPINO")
  # JAPANESE 'desu' or 'masu' suffix
  if sentence[-4:] == 'desu' or sentence[-4:] == 'masu':
    print("JAPANESE")
```
